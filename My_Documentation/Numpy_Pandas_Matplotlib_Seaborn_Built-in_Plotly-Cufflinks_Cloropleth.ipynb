### Converting string date column to timeStamp date

```python
from datetime import datetime
df_date = pd.DataFrame({'str': ['2015-12-10 17:40:00', '1915-12-30 08:40:30']})
df_date['timeStamp_1'] = pd.to_datetime(df_date['str'])
df_date['timeStamp_2'] = df_date['str'].apply(lambda date: datetime.strptime(date, '%Y-%m-%d %H:%M:%S')) # https://strftime.org/
print(df_date['str'][0], '\ttype:', type(df_date['str'][0]))
print(df_date['timeStamp_1'][0], '\ttype:', type(df_date['timeStamp_1'][0]))
print(df_date['timeStamp_2'][0], '\ttype:', type(df_date['timeStamp_2'][0]))
```

## Multi-index

```python
# Index Levels
outside = ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
inside = [1, 2, 3, 1, 2, 3]
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
df_1 = pd.DataFrame(np.random.randn(6, 2), index=hier_index, columns=['A', 'B'])
df_2 = pd.DataFrame(np.random.rand(6, 2), index=hier_index, columns=['A', 'B'])
df_1.index.names = ['Group', 'Num']
df_2.index.names = ['Group', 'Num']
df = pd.concat([df_1, df_2], keys=['df_1', 'df_2'], axis=1)
print(df)
```

## Merging, joining, and concatenating

### Concatenating

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
print(df1)
print(df2)
print(df3)
```

```python
pd.concat([df1, df2, df3], axis=0)
```

### Merging

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                         'C': ['C0', 'C1', 'C2', 'C3'],
                         'D': ['D0', 'D1', 'D2', 'D3']})
print(left)
print(right)
```

```python
pd.merge(left, right, how='inner', on=['key1', 'key2'])
```

### Joining

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
print(left)
print(right)
```

```python
left.join(right)
```

# Matplotlib

## Importing

```python
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

## Plotting

### Plotting a figure

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height
axes.plot(x, y)
axes.set_title('Larger Plot')
```

### Plotting figures, not using subplots

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height
axes1.plot(x, y)
axes1.set_title('Larger Plot')

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes2.plot(y, x)
axes2.set_title('Smaller Plot')
```

### Plotting figures, using subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,2))

axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second plot')

plt.tight_layout() # To avoid possible overlapping
```

### Saving a figure

```python
fig.savefig('my_figure.png', dpi=200)
```

### Inserting a legend

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend()
```

### Setting limits to the plot

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3)

ax.set_xlim([3,5])
ax.set_ylim([10, 100])
```

### Customizing a figure

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

# possible linestype options -, , -., :, steps
# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, y, color='purple', alpha=0.5, linewidth=3, linestyle='--',
        marker='o', markersize=12, markerfacecolor='red', markeredgewidth=2, markeredgecolor='black')
```
