# Multiple Linear Regression

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
df = pd.read_csv('50_Startups.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
print(X)
```

## Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
# You don't need to check for multicollinearity because we are going to test each
# model and select the best one anyways
```

```python
print(X)
```

## Splitting the dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Training the Multiple Linear Regression model on the Training set

```python
# Scikit learn library takes care of the dummy variable 
# trap and it also selects the best features for you too (directly in the equation)
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2) # sets precision of numpy arrays to 2
# Displaying the predicted results and the actual results.
# reshape is used to turn the arrays from horizontal to vertical
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), axis=1))
```

```python
regressor.predict([[1, 0, 0, 160000, 130000, 300000]])
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```
